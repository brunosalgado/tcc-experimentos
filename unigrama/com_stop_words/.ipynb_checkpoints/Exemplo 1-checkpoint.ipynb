#### Imports necessarios para funcionamento da aplicacao

```python
import re
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score, precision_score
from nltk import word_tokenize
from nltk.stem import RSLPStemmer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
from mlxtend.plotting import plot_confusion_matrix
from sklearn.metrics import confusion_matrix
```

#### Leitura do arquivo de dados

```python
df = pd.read_csv('../../base_a.csv')

df.head()
```

#### Efetua o tratamento dos dados

```python
le = LabelEncoder()

df.categoria = le.fit_transform(df.categoria)

df.head()
```

#### Função de tokenização

```python
stemmer = RSLPStemmer()

def token(text):
    text = re.sub('[0-9]|,|\.|/|$|\(|\)|-|\+|:|•', ' ', text)
    txt = word_tokenize(text.lower())
    return [stemmer.stem(word) for word in txt]
```

#### Lista de Stop Words

```python
stopwords_arquivo = pd.read_csv('../../stopwords.csv')

lista_stop_words = [ token(word)[0] for word in stopwords_arquivo.stopwords.values ]
```

#### Separa base de treino e teste

```python
X_train,X_test,y_train,y_test = train_test_split(
    df.nome,
    df.categoria,
    test_size = 0.3,
    random_state = 42
)
```

#### Define CountVectorizer

```python
count_vectorizer = CountVectorizer(
    tokenizer=token,
    ngram_range=(1,1),
    stop_words=lista_stop_words
)
```

#### Cria vocabulário

```python
X_train = count_vectorizer.fit_transform(X_train.values)
```

## Etapa de treinamento
#### Cria um classificador k-means

```python
kmeans2 = KMeans(
    n_clusters = 15,
#     init='k-means++',
#     max_iter=100,
#     n_init=1,
    init = 'random',
#     n_init=15,
#     max_iter=600,
    random_state=24
)
```

#### Treina o classificador

```python
kmeans2.fit(X_train.toarray(), y_train)
```

## Etapa de Teste

#### Efetua a predição

```python
Y_Pred = kmeans2.predict(count_vectorizer.transform(X_test).toarray())
```

#### Calcula a acurácia do classificador

```python
print('acuracia: {:.1f}'.format(accuracy_score(y_test, Y_Pred) * 100))
print('precisao: {:.1f}'.format(precision_score(y_test, Y_Pred, average="weighted") * 100))
```

#### Tabela com os itens classificados errados na base de teste

```python
d = []

for i in range(len(X_test)):
    au = count_vectorizer.transform([X_test.values[i]])
    pred = kmeans2.predict(au.toarray())
    if (y_test.values[i] != pred[0]):
        d.append(
            [
                X_test.values[i],
                le.inverse_transform([y_test.values[i]])[0],
                le.inverse_transform(pred)[0]
            ]
        )

df_error = pd.DataFrame(data=d,columns=['Nome','Categoria Original','Pred'])
df_error
```

#### Gráfico

```python
tsne = TSNE().fit_transform(PCA(n_components=25).fit_transform(X_train.todense()))
colors = ["#"+''.join([random.choice('0123456789ABCDEF') for j in range(6)]) for i in range(15)]

# Create plot
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(1, 1, 1)

for g in np.unique(y_train):
    ix = np.where(y_train == g)
    ax.scatter(tsne[:,0][ix], tsne[:,1][ix], c=colors[g], label = le.inverse_transform([g])[0], s = 8)

plt.title('Matplot scatter plot')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.axhline(color='b')
plt.axvline(color='b')
plt.show()
```

#### Termos por cluster

```python
order_centroids = kmeans2.cluster_centers_.argsort()[:, ::-1]
terms = count_vectorizer.get_feature_names()
for i in range(15):
    top_ten_words = [terms[ind] for ind in order_centroids[i, :10]]
    print("Cluster {}: {}".format(i, ', '.join(top_ten_words)))
```

#### Matriz de Confusão

```python
plot_confusion_matrix(
    conf_mat=confusion_matrix(y_test, Y_Pred),
    figsize=(20, 20),
    show_absolute=False,
    show_normed=True,
    colorbar=True,
    class_names=np.unique(le.inverse_transform(y_test))
)
```
